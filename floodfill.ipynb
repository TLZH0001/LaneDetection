```python
import numpy as np 
import cv2
import imutils
```

```python
cap = cv2.VideoCapture("/home/roar/Desktop/LaneDetection/training_1000.MOV")
cap.isOpened()
```

```python
while True:
    ret,frame = cap.read()
    if not ret:
        break
    img = frame.copy()
    img_hsv = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)
    frame = cv2.GaussianBlur(img_hsv, (5,5), 0)
    frame = cv2.GaussianBlur(frame, (5,5), 0)
    
    # cv2.imshow('image',blur)
    # key = cv2.waitKey(1) & 0xFF
    # if key == ord('q'):
    #     cv2.destroyAllWindows()
    #     break
    # continue


    #frame = imutils.resize(frame,width = 960)
    seed = (1079, 920)

    box = (200, 200)
    x = (seed[0] - box[0] - box[1], seed[0])
    y = (seed[1] - box[1], seed[1] + box[1])


    mean = np.mean(frame[x[0]:x[1], y[0]:y[1], :], axis = (0,1))
    # std = np.std(frame[x[0]:x[1], y[0]:y[1], :], axis = (0,1))
    # scale = 2.5
    # scale_std = std * scale
    # scale_std[2] = 200
    # scale_std[1] = 50
    # print(scale_std)
    # break
    scale_std = [50, 50, 200]
    # print(mean, '/n',std)

    frame[seed[0], seed[1]] = mean
    print(frame.shape)

    #crop_frame = frame[200:]
    #print(crop_frame.shape)

    
    mask = np.zeros((frame.shape[0] + 2, frame.shape[1] + 2)).astype(np.uint8)
    cv2.floodFill(frame, mask, seedPoint=seed, newVal=(255, 0, 0), \
        loDiff=tuple(scale_std), upDiff=tuple(scale_std), flags= cv2.FLOODFILL_FIXED_RANGE)
    cv2.circle(frame, seed, 2, (0, 255, 0), cv2.FILLED, cv2.LINE_AA)
    mask = mask * 255

    """ smooth the mask """
    kernel = np.ones((3,3), np.uint8)
    # mask = cv2.dilate(mask, kernel)
    # mask = cv2.erode(mask, kernel)
    
    """find the edge of mask"""
    #mask_copy = mask.copy()
    gx, gy = np.gradient(mask)
    edge = gy * gy + gx * gx
    edge[edge != 0.0] = 255.0
    edge = edge.astype(np.uint8)

    """ remove noise"""
    nb_components, output, stats, centroids = cv2.connectedComponentsWithStats(edge, connectivity=8)
    sizes = stats[1:, -1]
    nb_components = nb_components - 1
    min_size = 1000  #num pixels
    lane = np.zeros((output.shape))
    for i in range(0, nb_components):
        if sizes[i] >= min_size:
            lane[output == i + 1] = 255
            
    ''' smooth the edge'''
    lane2 = cv2.dilate(lane, kernel)

    
    cv2.imshow("floodfill result", frame)
    cv2.imshow('image',lane2)
    # cv2.imshow('frame_inv', frame_inv)
    # cv2.imshow('out', img_out)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        cv2.destroyAllWindows()
        break
```
